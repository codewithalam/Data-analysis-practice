```python
orders = [120, 150, 180, 200, 210, 220, 250, 260, 270, 280,
          290, 300, 310, 320, 350, 380, 400, 450, 500, 800]

expense_type = ['Food', 'Rent', 'Travel', 'Shopping']
amount = [8000, 15000, 3000, 6000]
```

```python
fig , axes = plt.subplots(1 , 2 , figsize=(15,4))

axes[0].hist(orders , bins=6 , edgecolor='black')
axes[0].set_title('Customer orders')
axes[0].set_xlabel('Orders')
axes[0].set_ylabel('No. of orders')

axes[1].pie(amount , labels=expense_type , autopct='%1.1f%%')
axes[1].set_title('Monthly expense')
plt.tight_layout()
plt.show()

# Watching two charts simultaneously, provide two insight at same time,  which save time and provide better summary for business perspective
```

```python
data = {
    'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'temperature': [22, 25, 28, 30, 26, 20, 24],
    'ice_cream_sales': [150, 200, 260, 300, 220, 140, 190]
}
df = pd.DataFrame(data)
```

```python
df.plot(x='temperature' , y='ice_cream_sales' , kind='scatter' , title='Ice-cream sales vs temperature' , legend=False)
plt.ylabel('Ice-cream sales')
plt.show()
```

```python
Annotations + Saving Charts 
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [50, 65, 60, 80]

plt.plot(months, revenue, marker='o', color='green')
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue (₹ thousands)')

plt.annotate('Best Month', 
           xy=('Apr',80),
           xytext=('Feb' , 75),
           arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.savefig('Monthly Revenue.png' , dpi=300 , bbox_inches='tight') 
plt.show()
```

```python
data = {
    'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'temperature': [22, 25, 28, 30, 26, 20, 24],
    'ice_cream_sales': [150, 200, 260, 300, 220, 140, 190]
}
df = pd.DataFrame(data)

df.plot(x='day' , y='ice_cream_sales' , kind='line' , title='Ice-cream sales' , legend=False)
plt.ylabel('No. of ice-cream sold')
plt.annotate('Peak sales', 
            xy=(3,300),
            xytext=(5,280),
            arrowprops=dict(facecolor='black' , arrowstyle='->'))
plt.savefig('ice_cream_sales.png' , dpi=300 , bbox_inches='tight')
plt.show()
```
